```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt
import time
place_name = "District 1, Ho Chi Minh City, Vietnam"
graph = ox.graph_from_place(place_name, network_type="drive")
origin_point = (10.7769, 106.6953)
destination_point = (10.7719, 106.6983)

origin_node = ox.nearest_nodes(graph, origin_point[1], origin_point[0])
destination_node = ox.nearest_nodes(graph, destination_point[1], destination_point[0])
start_time = time.time()
route_dijkstra = nx.shortest_path(graph, origin_node, destination_node, weight='length', method='dijkstra')
dijkstra_time = time.time() - start_time


start_time = time.time()
route_astar = nx.astar_path(graph, origin_node, destination_node, weight='length')
astar_time = time.time() - start_time
fig, ax = ox.plot_graph_route(graph, route_dijkstra, route_color='r', route_linewidth=4, node_size=0)

print(f"Thời gian chạy Dijkstra: {dijkstra_time:.6f} giây")
print(f"Thời gian chạy A*: {astar_time:.6f} giây")
```
