```python
import osmnx as ox
place_name = "District 1, Ho Chi Minh City, Vietnam"
graph = ox.graph_from_place(place_name, network_type="drive")
ox.plot_graph(graph, node_size=5, edge_linewidth=0.8, edge_color="#666666")
stats = ox.basic_stats(graph)
print(f"--- Kết quả phân tích tại {place_name} ---")
print(f"Số lượng nút giao (Nodes): {stats['n']}")
print(f"Số lượng cạnh/đoạn đường (Edges): {stats['m']}")
print(f"Chiều dài đường trung bình: {stats['edge_length_avg']:.2f} mét")
print(f"Mật độ giao lộ (Intersection density): {stats['intersection_count'] / stats['street_length_total']:.6f} per m")
```
