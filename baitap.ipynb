```python
import folium as fo
m = fo.Map(location=[10.7611662, 106.6684993], zoom_start=12)
location = [
    {"name": "UEH B", "coords":[10.7611662, 106.6684993], "group": "UEH", "color": "red"},
    {"name": "KTX UEH NCT", "coords": [10.760400799238809, 106.67197506792064], "group": "KTX", "color": "blue"},
    {"name": "Trạm Y Tế", "coords": [10.762033423864345, 106.66853850737485], "group": "TYT", "color": "green"},
    {"name": "Cửa Hàng Chuột", "coords": [10.761537799193537, 106.66787374620294], "group": "HS", "color": "orange"},
    {"name": "Munci Studio", "coords": [10.761787696140175, 106.66993137510633], "group": "STD", "color": "purple"}
]
layers ={
    "UEH": fo.FeatureGroup(name="UEH B"),
    "KTX": fo.FeatureGroup(name="KTX UEH NCT"),
    "TYT": fo.FeatureGroup(name="Trạm Y Tế"),
    "HS": fo.FeatureGroup(name="Cửa Hàng Chuột"),
    "STD": fo.FeatureGroup(name="Munci Studio")
}
for loc in location:
    fo.Marker(location=loc["coords"],popup=loc['name'],tooltip=loc["name"],icon=fo.Icon(color=loc["color"])).add_to(layers[loc["group"]])
for layer in layers.values():
    layer.add_to(m)
fo.LayerControl(collapsed=False).add_to(m)
m
```
