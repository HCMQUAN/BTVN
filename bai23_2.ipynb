```python
import folium
import osmnx as ox
import networkx as nx
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
import time
addresses = [
    "UEH A, Ho Chi Minh City",
    "Trường Đại học Bách khoa - Đại học Quốc gia TP.HCM",
    "Sân vận động Thống Nhất",
    "Ben Thanh Market, Ho Chi Minh City",
    "Saigon Notre Dame Cathedral",
    "Independence Palace Ho Chi Minh City",
    "Saigon Zoo",
    "Vinh Nghiem Pagoda",
    "Bệnh viện Đại học Y Dược TP.HCM",
    "Bệnh viện Quân Y 7A",
    "Bệnh viện Bệnh Nhiệt đới"
]
geolocator = Nominatim(user_agent="map_project")
locations = []
for addr in addresses:
    location = geolocator.geocode(addr)
    if location:
        locations.append({
            "name": addr,
            "coords": (location.latitude, location.longitude)
        })
    time.sleep(1)
center = locations[0]
m = folium.Map(location=center["coords"], zoom_start=13)
layer_points = folium.FeatureGroup(name="Địa điểm")
layer_lines = folium.FeatureGroup(name="Khoảng cách")
for loc in locations:
    distance = geodesic(center["coords"], loc["coords"]).km
    folium.Marker(
        location=loc["coords"],
        popup=loc['name'],
        tooltip=loc["name"],
        icon=folium.Icon(color="blue" if loc == center else "green")
    ).add_to(layer_points)
    folium.PolyLine(
        locations=[center["coords"], loc["coords"]],
        color="red",
        weight=2,
        opacity=0.6,
        popup=f"{distance:.2f} km"
    ).add_to(layer_lines)
layer_points.add_to(m)
layer_lines.add_to(m)
folium.LayerControl().add_to(m)
m
```
