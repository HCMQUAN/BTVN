```python
import folium
center = [10.7611662, 106.6684993]
m = folium.Map(location=center, zoom_start=13)
folium.Circle(
    location=center,
    radius=3000,
    color="green",
    fill=True,
    fill_opacity=0.2,
    popup="3 km"
).add_to(m)
folium.Circle(
    location=center,
    radius=5000,
    color="orange",
    fill=True,
    fill_opacity=0.2,
    popup="5 km"
).add_to(m)
folium.Circle(
    location=center,
    radius=10000,
    color="red",
    fill=True,
    fill_opacity=0.2,
    popup="10 km"
).add_to(m)
folium.Marker(
    location=center,
    popup="Trung tâm (UEH)",
    icon=folium.Icon(color="blue")
).add_to(m)
m
```
